```python
import pandas as pd 
```

```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
people
```

```python
df = pd.DataFrame(people)
```

```python
df
```

### Indexes are something that act like row label

##### Pandas doesn't enforces the index to be unique, and sometimes it won't be but most of the time it is

**Let us make the email as the indexes**

```python
### To see all the email addresses
df['email']
```

### Setting email addresses as the index of the DataFrame

```python
df.set_index('email', inplace=True)
```

```python
df
```

```python
df.index 
```

**Here the index has the same name as the column**

```python
df.index.name 
```

Let's set the name of the index as None

```python
df.index.name = None 
```

```python
df 
```

```python
df.index 
```

As email is the index of the DataFrame, we can use it to fetch a row uniquely using loc

```python
df.loc['CoreyMSchafer@gmail.com']
```

Fetching the column 'last' using loc and index

```python
df.loc['CoreyMSchafer@gmail.com','last']
```

Resetting the index

```python
df.index.name = 'email'
df.reset_index(inplace=True)
```

```python
df 
```
